## Data. 
#### From where and how are you collecting your data? If appropriate, provide a link to your data sources.
We will be collecting our data from https://www.winemag.com. They have 215,395 reviews of wines including price.<br>
##### These are the fields of each review we plan on scraping:
<br>Url - Where the review is hosted.
<br>Title - Name of the review.
<br>Points - Number of points given in the review. On a 100 point scale.
<br>Description - The taster's review of the wine.
<br>Price - Price of the wine. We think it's all in dollars.
<br>Variety - Type of wine.
<br>Appellation - a name or title of the wine.
<br>Winery - Location the wine was produced.
<br>Alcohol - Percent of achohol.
<br>Bottle Size - Size of the wine bottle. It might all be in millilitres.
<br>Category - Category of the wine.
<br>Importer - Name of the importer.
<br>Date Published - Date the review was published.
<br>User Avg Rating - Rating of the review.
<br>Taster - Name of the reviewer.

## Exploratory Analysis.
#### Which methods and visualizations are you planning to use to look at your dataset?
Scatterplots and correlations across the features.
Partial correlations to investigate multicollinearity.
Investigating the principal components and factor analysis loadings on the reviews to find themes.<br>
Using Penalised linear models (lasso) to identify important features to feed into trees and svms.<br>
Exploring different clustering methods to find patterns and then profiling on these clusters to gain insight into our data.<br>

## Analysis Methodology. 
#### How are you planning to analyze your data?

What are the key drivers for wine points and wine price? Penalised linear models (elastic net) and trees.<br>
Which wines are systematically over or under valued in terms of price? Penalised linear models, random forests, gradient<br> boosting machines, k-NNs, SVMs, and Neural Networks.<br>
What attributes per wine should you use to market them? PCA, FA, SEMs, and Bayesian Networks.<br>
How similar are different wines varieties? Clustering and profiling.<br>
